```python
from sympy import *
```

**Объявление функции**

```python
x = Symbol('x')
function = (sin(x)/2 + cos(x)/2) ** 2 * sin(x)
function
```

**Упрощенная функция**

```python
func =  trigsimp(function)
func
```

**Построение графика**

```python
grafic = plot(function)
```

**Нахождение корней**

```python
func_root=sorted(solve(func))
func_root
```

**Производная**

```python
derivative = diff(function,x)
derivative1 = trigsimp(derivative)
derivative1
```

**График 1 периода**

```python
period_schedule = plot(func,derivative1,(x,0,2*pi))
```

**Корни производной**

```python
deriv_root = sorted(solve(derivative))
deriv_root
```

**Периодичность функции**

```python
interval=periodicity(function,x)
interval
```

**Интервалы, на которых функция возрастает и убывает**

```python
solve([x >=0, x <= periodicity(func, x), func], x)
```

**Вершина функции**

```python
for derivative in deriv_root:
    point = function.subs(x, derivative)
    print(f'Координаты вершины:\n x = {round(derivative.evalf(),2)},\n y = {round(point.evalf(),2)}')
```

**Определить промежутки, на котором f > 0** 


```python
solveset(func>0, x, S.Reals)
```

**Определить промежутки, на котором f < 0**

```python
solveset(func<0, x, S.Reals)
```
